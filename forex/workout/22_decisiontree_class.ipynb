```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

import allinone as aio

%matplotlib inline
warnings.filterwarnings('ignore')
pd.set_option("display.float_format", "{:.3f}".format)
pd.set_option('display.max_columns', 999)
plt.style.use('ggplot')
```

```python
df_dummy = pd.read_csv('forex_dummy.csv')
```

#### Scale

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df_dms = df_dummy[['buy','ind_40_1','ind_40_2','ind_40_3','ind_40_4','ind_40_5','ind_40_6','ind_217_0','ind_217_1','ind_217_2','ind_217_3','ind_217_4','ind_217_5']]
df_scale = df_dummy.drop(['buy','ind_40_1','ind_40_2','ind_40_3','ind_40_4','ind_40_5','ind_40_6','ind_217_0','ind_217_1','ind_217_2','ind_217_3','ind_217_4','ind_217_5'], axis=1)

#StandardScaler
scale_col = pd.DataFrame(StandardScaler().fit_transform(df_scale),
            columns=df_scale.columns)
dfScale = pd.concat([df_dms, scale_col], axis=1)
```

## Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
y = df_dummy['buy']
X = df_dummy.drop(['90_target','buy'], axis=1)
model = DecisionTreeClassifier()
```

```python
aio.confusion(X, y, model)
```

```python
dfModel = pd.DataFrame(columns = ['cross-score', 'cross-train', 'train_score', 'test_score', 'precision', 'recall', 'f1-score', 'auc-roc', 'auc-pr'])
```

```python
y_scale = dfScale['buy']
X_scale = dfScale.drop(['90_target','buy'], axis=1)
model = LogisticRegression()

dfModel = dfModel.append(aio.modelframe(X_scale, y_scale, model, 'logistic'))
```

```python
y_scale = dfScale['buy']
X_scale = dfScale.drop(['90_target','buy'], axis=1)
model = DecisionTreeClassifier()

dfModel = dfModel.append(aio.modelframe(X_scale, y_scale, model, 'tree'))
```

### Parametreler

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, KFold
```

```python
y_scale = dfScale['buy']
X_scale = dfScale.drop(['90_target','buy'], axis=1)

dtree = DecisionTreeClassifier()
params = {'min_samples_split': [2, 4, 8, 16, 32, 48, 64], 'max_depth': [2, 4, 8, 16, 32, 48, 64], 'min_samples_leaf': [2, 8, 16, 32, 48, 64]}
```

```python
k_fold = KFold(n_splits = 10, shuffle = True, random_state = 42)
gridcv_results = GridSearchCV(estimator = dtree, param_grid = params, cv = k_fold, n_jobs = -1, verbose = 0).fit(X_scale, y_scale)
```

```python
gridcv_results.best_params_
```

```python
model = DecisionTreeClassifier(max_depth=16, min_samples_leaf=2, min_samples_split=4)

dfModel = dfModel.append(aio.modelframe(X_scale, y_scale, model, 'tree_p1'))
```
